Задача 1

Даны значения зарплат из выборки выпускников:

100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150

Используя только встроенные питоновские функции и структуры данных (т.е. без библиотек numpy, pandas и др.) посчитать среднее арифметическое, смещённую и несмещённую оценки дисперсии, среднее квадратичное отклонение для данной выборки.

Можно затем посчитать те же значения с использованием библиотек, чтобы проверить себя.

```python
import numpy as np

a = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150]

sum = 0
for i in range(len(a)):
    sum = sum + a[i] 
a_mean = sum / len(a)
print(f'Среднее арифметическое:',round(a_mean,3))

b = 0
for i in range(len(a)):
    b = b + (a[i] - a_mean)**2
a_ov = b / len(a)
print(f'Смещеная оценка дисперсии',round(a_ov,3))

a_uv = b / (len(a) - 1)
print(f'Несмещенная оценка дисперсии:',round(a_uv,3))

std = a_ov ** 0.5
print(f'Среднее квадратичное отклонение:',round(std,3))

print(a_mean == np.mean(a))
print(a_ov == np.var(a))
print(a_uv == np.nanvar(a))
print(std == np.std(a))
```

Задача 2

Для выборки из задачи 1 найти (также без использования библиотек):
1. медиану, первый и третий квартили, интерквартильное расстояние,
2. выборсы в выборке (используя для этого метод как при построении "усов" из boxplot).

Возможные неоднозначности в вычислении квантилей можно разрешать любым способом.

```python
# Отсортируем список
n = 1
while n < len(a):
    for i in range(len(a) - n):
        if a[i] > a[i + 1]:
            a[i], a[i + 1] = a[i + 1], a[i]
    n +=1
print(a)

# Найдем медиану, 1-й и 3-й квартили и интерквартильное рассояние
a_median = a[int((len(a) + 1) / 2)]
a_25 = a[int((len(a) + 1) * 0.25)]
a_75 = a[int((len(a) + 1) * 0.75)]
a_inter = a_75 - a_25

print(f'Медиана равна:',a_median)
print(f'1-й квартиль равен:',a_25)
print(f'3-й квартиль равен:',a_75)

# Найдем выбросы в выборке
a_vmin = a_25 - a_inter
a_vmax = a_75 + a_inter
for i in range(len(a)):
    if a[i] < a_vmin | a[i] > a_vmax:
        print(f'Элемент',a[i],'является выбросом')
```

Задача 3

Какова вероятность, что он учится: a) на факультете A? б) на факультете B? в) на факультете C?

Замечание: да, эта задача не на тему описательных статистик, но тема важная, и её стоит иногда освежать в памяти

```python
P_fA = 1 / 4 * 0.8
P_fB = 1 / 4 * 0.7
P_fC = 2 / 4 * 0.9

print(f'a) Вероятность того, что студент учится на факультете А равна:',P_fA)
print(f'б) Вероятность того, что студент учится на факультете B равна:',P_fB)
print(f'в) Вероятность того, что студент учится на факультете C равна:',P_fC)
```
