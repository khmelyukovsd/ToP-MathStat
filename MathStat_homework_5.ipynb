Известно, что генеральная совокупность распределена нормально
со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
если выборочная средняя M = 80, а объем выборки n = 256.

```python
import numpy as np
from scipy import stats

n = 256
gamma = 0.95
M = 80
std = 16

t_p= stats.norm.ppf(1 - gamma / 2) # нормируемая величина
E = t_p * std / n**0.5 #  доверительная оценка

print(f'Истинное значение величины X:\n', round(M - E, 3),' < X <',round(M + E, 3))
```

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,
получены опытные данные:
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,
оценить истинное значение величины X при помощи доверительного интервала, покрывающего это
значение с доверительной вероятностью 0,95.

```python
import numpy as np
from scipy import stats

a = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
gamma = 0.95
var = np.var(a)
std = var**0.5

t_p= stats.norm.ppf(1 - gamma / 2) # нормируемая величина
E = t_p * std / len(a)**0.5 #  доверительная оценка

print(f'Истинное значение величины X:\n', round(np.mean(a) - E, 3),' < X <',round(np.mean(a) + E, 3))
```

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

```python
from scipy import stats

mu = 17 # - нулевая гипотеза. mu > 17мм - альтернативная гипотеза
alpha = 0.05
n = 100
var = 4
mean = 17.5
std = var**0.5

T = (mean - mu) / (std / (n**0.5)) # наблюдаемое значение критерия
print(f'Наблюдаемое значение критерия;',T)

#критическая область
s = stats.norm.ppf(1 - (2 * alpha) / 2)
print(f'Критическая область:',round(s,3))

if T < s:
    print(f'Принимаем нулевую гипотезу, о том, что средний диаметр подшипника равен 17мм')
else:
    print(f'Принимаем альтернативную гипотезу, о том, что средний диаметр подшипника больше 17мм')
```

Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

```python
import numpy as np
from scipy import stats

mu = 200 # - нулевая гипотеза. mu > 17мм - альтернативная гипотеза
a = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
alpha = 0.99
std = np.var(a)**0.5

T = (np.mean(a) - mu) / (std / (len(a)**0.5)) # наблюдаемое значение критерия
print(f'Наблюдаемое значение критерия;',round(T,3))

#критическая область
t = stats.t.ppf(1 - alpha / 2, len(a)-1)
print(f'Критическая область:',round(t,3))

if T < t:
    print(f'Принимаем нулевую гипотезу, о том, что средний вес пачки печенья равен 200г')
else:
    print(f'Принимаем альтернативную гипотезу, о том, что средний вес пачки печенья больше 200г')
```
